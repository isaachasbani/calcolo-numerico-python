```python
x=  graf2["x"];
y=  graf2["y"];
a=  graf1["A"];
b=  graf1["B"];
a= a.iloc[0];
b= b.iloc[0];
n=35;
z=0;
r=0;
h=(b-a)/n
```

```python
#metodo dei rettangoli

for i in range(n):
    z= z + y.iloc[i]

for t in range(1, n+1):
    r= r+ y.iloc[t]

    

```

```python
# calcolo della somma per eccesso e per difetto  
    
s= h * z
S= h * r

print(s)
print(S)
```

```python
#calcolo dell'errore approssimato

err= (b-a)**2/(2*n) * y.max()
print("Il risultato dell'integrale calcolato con il metodo dei rettangoli  compreso tra:", s, "e",  S, "con un errore approssimato <=", err)
```

```python
#metodo dei trapezi
fa= y.iloc[0]
fb= y.iloc[35]
somma=0

for q in range (1, n):
    somma = somma + y.iloc[q]


result= h* ((fa+fb)/2 + somma)
```

```python
print("Il risultato dell'integrale calcolato con il metodo dei trapezi  ", result)
```
